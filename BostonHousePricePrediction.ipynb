```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.layers import Dropout
```

```python
X = df.drop('MEDV', axis=1)
y = df['MEDV']
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, y, test_size=0.3)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import BatchNormalization
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(13,)),
    BatchNormalization(),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dense(1)
])
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(loss='mse',
              optimizer=Adam(lr=0.01),
              metrics=['mae'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=500)
```

```python
hist = model.fit(X_train, Y_train,
          batch_size=32, epochs=5000,
          validation_data=(X_val, Y_val))
```

```python
plt.plot(hist.history['mae'], label='train mae')
plt.plot(hist.history['val_mae'], label='val mae')
plt.xlabel('epoch')
plt.ylabel('mae [1000$]')
plt.legend(loc='best')
plt.ylim([0,50])
plt.show()
```

```python
model.evaluate(X_test, Y_test)[1]
```

```python
y_pred = model.predict(X_test).flatten()

p = np.polyfit(Y_test, y_pred, 1)
f = np.poly1d(p)

plt.scatter(Y_test, y_pred)
plt.plot(Y_test, f(Y_test), color = "r")
plt.xlabel("True Value")
plt.ylabel("Predicted Value")
plt.show()
```
